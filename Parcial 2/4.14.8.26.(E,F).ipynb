```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Punto E

def funcion(x):
  return x**3

def suma_riemmann_x3(n):
  return 4*(1 - (2/n) + (1/n)**2)

def suma_riemann(f, a, b, n):
    delta_x = (b - a) / n
    x = np.linspace(a + delta_x, b, n)
    y = f(x)
    integral = np.sum(y) * delta_x
    return integral

print(suma_riemmann_x3(30))
print(suma_riemann(funcion, 0,2,30))
```

```python
# Punto F
# Valor de la integral exacta va a ser: 1/4 *[x^4]0->2 = 1/4 * [2^4 - 0^4] = (2^4)/(2^2) = 2^2 = 4
# Por lo tanto Iexacta = 4

def error(n):
  Iexacta = 4
  return np.abs(Iexacta - suma_riemmann(n))
N = np.linspace(30,400,100, endpoint = False)
plt.plot(N,error(N), color = (0.7159,0.1333,0.0118))
plt.xlabel("Valor de n")
plt.ylabel("Error")
plt.grid()
plt.title("Error vs valor de n")
plt.legend(["Ïµ(n)"])
plt.show()
```
