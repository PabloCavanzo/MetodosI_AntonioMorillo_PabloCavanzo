```python
fileL = '/content/drive/MyDrive/Colab/MetodosI_AntonioMorillo_PabloCavanzo/MinimosLineal.txt'
fileC = '/content/drive/MyDrive/Colab/MetodosI_AntonioMorillo_PabloCavanzo/MinimosCuadratico.txt'
```

```python
def fL(x,a0,a1):
  return a0+a1*x
def fC(x,a0,a1,a2):
  return a0+a1*x+a2*x**2
```

```python
dataL = pd.read_csv(fileL, sep=' ', header=None)
dataL.columns = ['x', 'y']
xL = np.array(dataL['x'])
yL = np.array(dataL['y'])
dataC = pd.read_csv(fileC, sep=' ', header=None)
dataC.columns = ['x', 'y']
xC = np.array(dataC['x'])
yC = np.array(dataC['y'])
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
curvaL = spo.curve_fit(fL,xL,yL)
ax[0].plot(xL,yL,'o')
ax[0].plot(xL,fL(xL,*curvaL[0]))
ax[0].grid()
ax[0].set_title('Ajuste Lineal')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
curvaC = spo.curve_fit(fC,xC,yC)
ax[1].plot(xC,yC,'o')
ax[1].plot(xC,fC(xC,*curvaC[0]))
ax[1].grid()
ax[1].set_title('Ajuste Cuadratico')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
plt.show()
print('Lineal:')
print('f(x) = {} + {}x'.format(round(curvaL[0][0],4),round(curvaL[0][1],4)))
print('\nCuadratico:')
print('f(x) = {} + {}x + {}x^2'.format(round(curvaC[0][0],4),round(curvaC[0][1],4),round(curvaC[0][2],2)))
```
