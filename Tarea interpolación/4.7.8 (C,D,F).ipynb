```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return np.sqrt(np.tan(x))
```

```python
x = np.linspace(0.1,1.1,100)
```

```python
def Derivada_progresiva(f,x,h=0.01):
  return (1/(2*h))*(4*f(x + h) - 3*f(x) - f(x + 2*h))
Derivada_progresiva(f,x)
# Punto C
```

```python
def Derivada_central(f,x,h=0.01):
  return (f(x + h) - f(x - h))/(2*h)
Derivada_central(f,x)
# Punto D
```

```python
def Df(x):
  cos = np.cos(x)
  tan = np.tan(x)
  return 1 / (2*(cos**2)*(tan**(1/2))) # Fundamentacion teorica de derivada exacta en PDF
Df(x)
# Punto E
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)
axs[0,0].plot(Derivada_progresiva(f,x,h=0.01), color = 'r')
axs[0,0].set_title('Derivada progresiva')
axs[0,0].legend(['Derivada progresiva'])
axs[0,0].grid()
axs[1,0].plot(Derivada_central(f,x,h=0.01), color = 'b')
axs[1,0].set_title('Derivada central')
axs[1,0].legend(['Derivada central'])
axs[1,0].grid()
axs[0,1].plot(f(x), color = 'g')
axs[0,1].set_title('Funcion')
axs[0,1].legend(['f(x)'])
axs[0,1].grid()
axs[1,1].plot(x,Df(x))
axs[1,1].set_title('Derivada exacta')
axs[1,1].legend(['Derivada exacta'])
axs[1,1].grid()
plt.subplots_adjust(hspace=0.5)
plt.show()
# Grafica de puntos C, D y E juntas
```

```python
plt.plot(x,Derivada_progresiva(f,x), color = 'r')
plt.plot(x,Derivada_central(f,x), color = 'b', linestyle = '--')
plt.plot(x,Df(x), color='y', linestyle = ':')
plt.plot(x,f(x), color = 'g')
plt.legend(['Derivada progresiva','Derivada central', 'Derivada exacta', 'f(x)'])
plt.grid()
plt.show()
# Grafica de puntos C, D y E juntas
```

```python
def error_nodal(D_real, D_aprox):
  return np.abs(D_real - D_aprox)
# Punto F
```

```python
fig, axs = plt.subplots(nrows=2)
axs[0].plot(x,error_nodal(Df(x), Derivada_progresiva(f,x)), color = 'r')
axs[1].plot(x,error_nodal(Df(x), Derivada_central(f,x)), color = 'b')
axs[0].set_title('Error derivada progresiva')
axs[0].grid()
axs[1].set_title('Error derivada central')
axs[1].grid()
plt.subplots_adjust(hspace=0.5)
plt.show()
# Punto F Grfica
```

Como se puede ver por las grficas, ambos resultados tienen efectivamente el mismo orden de presicin. Pues, ambos resultados tienen un orden de presicion de 10^-3.
