```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

def Lagrange(x,X,i):
  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L

def Interpolate(x,X,Y):
  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*Lagrange(x,X,i)
  return Poly

def funcion(x):
  return np.sqrt(x)

x = sym.Symbol("x",real=True)

X = np.array([100,121,144])
Y = np.array([10,11,12])
coords_x = np.linspace(0,X[-1],100)
coords_y = Interpolate(coords_x,X,Y)
y_real = funcion(coords_x)
error_y = np.abs(coords_y-y_real)

fig, ax = plt.subplots()
ax.plot(coords_x,y_real,color="red",label="√x")
ax.plot(coords_x,coords_y,label="Polinomio")
ax.plot(coords_x,error_y,color="red",linestyle="--",label="Error")
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
ax.set_title("Gráfica ente 0 y 144")
ax.legend()
plt.show()


error = np.abs(funcion(114)-Interpolate(114,X,Y))
print("Polinomio: ", sym.expand(Interpolate(x,X,Y)))
print("El error al calcular √114 es:",error)
```

```python
coords_x2 = np.linspace(0,1000,100)
coords_y2 = Interpolate(coords_x2,X,Y)
y_real2 = funcion(coords_x2)
error_y2 = np.abs(coords_y2-y_real2)

fig, ax = plt.subplots()
ax.plot(coords_x2,y_real2,color="red",label="√x")
ax.plot(coords_x2,coords_y2,label="Polinomio")
ax.plot(coords_x2,error_y2,color="red",linestyle="--",label="Error")
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
ax.set_title("Gráfica ente 0 y 1000")
plt.legend()
plt.show()
```
