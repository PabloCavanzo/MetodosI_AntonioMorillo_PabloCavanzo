```python
import matplotlib.pyplot as plt
import numpy as np
import math

dias_totales = 365

def probabilidad_cumpleanos(n):
    prob = 1.0
    for i in range(n):
        prob *= (dias_totales - i) / dias_totales
    return prob
# Lo que hace esto es calcular 365*364*...*(365 - n + 1)/(365^n), lo cual es igual a [365!/((365-n)!*(365^n))] (fórmula inducida).
# Es decir, genera la misma respuesta evitando costo computacional, calculando lo necesario únicamente

n_values = np.arange(1, 81)
probabilities = [probabilidad_cumpleanos(n) for n in n_values]

plt.figure(figsize=(10, 6))
plt.plot(n_values, probabilities, color='c',marker='o')
plt.title("Probabilidad de que las fechas sean diferentes para n personas")
plt.xlabel("Número de personas (n)")
plt.ylabel("Probabilidad")
plt.grid()
```
