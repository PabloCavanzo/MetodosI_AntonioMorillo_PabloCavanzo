```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,1000)
```

```python
# sea y = S(N, x)/(Kb*N)
y = -1*(x*np.log(x) + (1 - x)*np.log(1 - x))
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('$S(N, x)/(Kb*N) vs. (x = n_1/N)$')
plt.show()
```
